```python
import pandas as pd
from astropy.io import fits
import math 
from astropy.wcs import WCS


def omega(bmaj,bmin):
    return np.pi*bmin*bmaj/(4*np.log(2))

def convert_arcsec_to_pixels(Image_PATH,spacial_width,header=None):
    '''Convert arcsec widths to pixels numbers.'''
    # Load the image header
    if header == None:
        print('Loading the image header')
        header = fits.open(Image_PATH)[0].header
    else:
        print('Using the provided header')
        header = header
    # Read the header parameters
    arcseconds_per_pixel = header['CDELT1']*3600*-1
    beam_size_arcseconds = header['BMAJ']*3600
    oversampled_spacial_width = (spacial_width**2 + beam_size_arcseconds**2)**0.5
    return oversampled_spacial_width/arcseconds_per_pixel

def correc_for_area_profound(row,sigma):
    ## we are making a correction as we are, like profound calculating the flux density from the segmentation, which is limited to the threshold level.
    # this is automatically done in pybdsf as it calculates all properties from the modelled gaussians. Here we are correcting for the faintest sources.
    omega_ = omega(sigma*2,sigma*2)
    if row.N100 <= 10 and row.flux <= 1E-2:
        # equatino for a 2d gaussian 
        x, y = np.meshgrid(np.arange(0, 100, 1), np.arange(0, 100, 1))
        gaussian = row.flux * np.exp(-((x-50)**2 + (y-50)**2)/(2*sigma**2))
        return gaussian.sum()/omega_
    
    else:
        return row.flux/omega_

def pf_peakflux(row,image):
    x = row.xmax
    y = row.ymax
    return image[int(y),int(x)]

def get_profound_catalogue(pf_path, image, bmaj_p):
    pf_cat = pd.read_csv(pf_path)
    pf_cat['Peak_flux'] = pf_cat.apply(lambda row: pf_peakflux(row,image),axis=1)
    pf_cat['flux_corr'] = pf_cat.apply(lambda row: correc_for_area_profound(row,bmaj_p/2),axis=1)
    return pf_cat

def get_pybdsf_catalogue(pybdsf_path):
    pybdsf_cat = fits.open(pybdsf_path)[1].data
    return pybdsf_cat


def xy_to_radec(fits_filename, x_coords, y_coords):
    # Open the FITS file and extract the WCS information from the header
    with fits.open(fits_filename) as hdul:
        wcs = WCS(hdul[0].header)
    
    # Create a list to store the (RA, Dec) pairs
    radec_coords = []
    ra_coords = []
    dec_coords = []
    # Convert each (x, y) pair to (RA, Dec)
    for x, y in zip(x_coords, y_coords):
        # The pixel coordinates are 0-based, so we add 1 to them
        pixel_coords = np.array([[x+1,y+1,0,0]], dtype=np.float64)

        # Use the WCS object to perform the transformation
        sky_coords = wcs.all_pix2world(pixel_coords[:,0],pixel_coords[:,1],pixel_coords[:,2],pixel_coords[:,3],0)
    
        # Extract the RA and Dec from the result
        ra_deg, dec_deg, pol, freq = sky_coords
        ra_coords.append(ra_deg)
        dec_coords.append(dec_deg)

    return ra_coords, dec_coords


def grehls_error(counts,confidence):
    e_l = confidence*math.sqrt(counts+1) +(confidence**2 -1)/(4)
    e_u = confidence*math.sqrt(counts+1) +(confidence**2 +2)/(3)
    return e_u, e_l

def nS(df_flux,bins,area):
    ns = []
    centers = []
    ns_ue_list = []
    ns_le_list = []
    for i in range(len(bins)-2):
        bin_width = bins[i+1] - bins[i]
        bin_center = (bins[i] + bin_width) # S
        centers.append(bin_center)
        counts = len(df_flux[(df_flux> bins[i]) & (df_flux < bins[i+1])])
        #counts_Plus_one = len(Source_count_Data[(Source_count_Data['Flux'] > fluxbins[i+1]) & (Source_count_Data['Flux'] < fluxbins[i+2])])
        print(counts)
        dN = counts
        dN_eu, dN_el = grehls_error(counts,confidence=1)
        
        ds = bin_width
        n1 = (dN/ds)
        #print(n1)
        n2 = (bin_center**(2.5))
        #print(n2)
        #print(area)#
        if ns == 0:
            continue

        ns.append(n1*n2/area)
        try:
            ns_ue = ((dN_eu/dN))*(dN/ds)*n2/area
            ns_le = ((dN_el/dN))*(dN/ds)*n2/area
        except ZeroDivisionError:
            ns_ue = 0
            ns_le = 0
        ns_ue_list.append(ns_ue)
        ns_le_list.append(ns_le)

    return ns,centers, ns_ue_list, ns_le_list

def sq_deg_to_str(area):
    return area*4*np.pi/((360*360)/np.pi)

def get_BDSF_cat(bdsf_gaul_PATH):
    bdsf_gaul = fits.open(bdsf_gaul_PATH)[1].data
    bdsf_gaul_df = pd.DataFrame(bdsf_gaul)
    bdsf_gaul_df['RA'] = bdsf_gaul_df['RA'].astype(float)
    bdsf_gaul_df['DEC'] = bdsf_gaul_df['DEC'].astype(float)
    bdsf_gaul_df['Total_flux'] = bdsf_gaul_df['Peak_flux'].astype(float)
    bdsf_gaul_df = bdsf_gaul_df[['RA','DEC','Total_flux']]
    return bdsf_gaul_df

```
